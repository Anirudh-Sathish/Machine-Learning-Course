```python
# Importing neccesary libraries
import pandas as pd
import matplotlib.pyplot as plt 
```

```python
#Reading the csv file regarding auto 
df = pd.read_csv("Auto.csv")
```

```python
# To find average of cylinders using groupby 
df_cylinder_grouping = df.groupby("cylinders")
averageMPG = df_cylinder_grouping["mpg"].mean()
print(averageMPG)
```

```python
# Plotting Scatter plot between mpg and displacement
plt.scatter(x=df["mpg"],y=df["displacement"])
```
