```python
import numpy as np 
from scipy.stats import multivariate_normal
import random 
import matplotlib.pyplot as plt 
```

```python
mean = [1,2,1]
#mean = np.array(mean).transpose
#mean = np.asmatrix(mean)
#print(mean.shape)

cov = [[5,0.8,-0.3],[0.8,3,0.6],[-0.3,0.6,4]]
cov = np.array(cov)
cov = np.asmatrix(cov)
```

```python
result = []
result = np.random.multivariate_normal(mean,cov,5000).T
# Now result contains the required distribution 
# Now we have to obtain 300 3 dimensional vectors from this distribution 
# let us make the selection randomly 
selection = random.sample(range(0, len(result[0])), 300)
# The above values are the ones we are going to select
vectors = []
vectors_x1 = []
vectors_x2 = []
vectors_x3 = []
for i in range(0,len(selection)):
    vector = []
    x1 = result[0][selection[i]]
    x2 = result[1][selection[i]]
    x3 = result[2][selection[i]]
    vectors_x1.append(x1)
    vectors_x2.append(x2)
    vectors_x3.append(x3)
    vector = [x1,x2,x3]
    vectors.append(vector)


# Obtained 300 3Dimensional Vectors 
```

```python
# Moving on to the next part of printing scatter plots 
# plotting x1 vs x2 
plt.scatter(x = vectors_x1,y = vectors_x2)
plt.show()
```

```python
# plotting x1 vs x3 
plt.scatter(x = vectors_x1,y = vectors_x3)
plt.show()
```

```python
# plotting x2 vs x3 
plt.scatter(x = vectors_x2,y = vectors_x3)
plt.show()
```

The relationships gathered from the above plots are : 
1. The distributions seem very simmilar 

```python
vectors = np.array(vectors)
mean_new = np.mean(vectors ,axis = 0)
print(mean_new)
```

```python
# Covariance matrix 
cov_new = np.cov(vectors.transpose())
print(cov_new)
```
