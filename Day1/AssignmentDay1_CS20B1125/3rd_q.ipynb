```python
import numpy as np 
from PIL import Image
import matplotlib.pyplot as plt  
```

```python
img = Image.open("trial1.jpg")
```

```python
mat1 = np.asarray(img)
#extract red channel
red_channel = mat1[:,:,0]
plt.imshow(red_channel)
```

```python
patch_32 = red_channel[400:432,200:232]
plt.imshow(patch_32)
print(patch_32.shape)
```

```python
# To find maximum values 
for i in range(len(patch_32)):
    # Maximum value of each row
    print("Maximum pixel value of Column ", i+1)
    print(max(patch_32[:,i]))
    print("Maximum pixel value of Row ", i+1)
    print(max(patch_32[i,:]))
```

```python
# Applying the cos change 
new_patch = []
for column in patch_32:
    rows = []
    for row in column:
        val = np.cos(row).tolist()
        rows.append(val)
    new_patch.append(rows)
plt.imshow(new_patch)
```
